```python
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets

def plot_knn_bound(X, y, scaler=None, n_neighbors=10, xlim=(-15, 15), ylim=(-20, 20)):
    # step size in the mesh
    h = 0.05

    # Create color maps
    cmap_light = ListedColormap(['C0', 'orange', 'cyan', 'green'][:np.unique(y).shape[0]])
    cmap_bold = ['C0', 'orange', 'c', 'darkgreen'][:np.unique(y).shape[0]]

    x_min, x_max = xlim
    y_min, y_max = ylim
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    grid = np.c_[xx.ravel(), yy.ravel()]

    X_scaled = X # if scaler is None
    if scaler is not None:
        grid = scaler.transform(grid)
        X_scaled = scaler.transform(X)

    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(n_neighbors, algorithm='brute')
    clf.fit(X_scaled, y)

    Z = clf.predict(grid)

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y,
                    palette=cmap_bold, alpha=1.0, edgecolor="black")
    plt.xlabel('x_1')
    plt.ylabel('x_2')
    plt.title('   {}-NN'.format(n_neighbors))
    plt.grid()
    plt.show()
```

   $1$-$NN$  $10$-$NN$

```python
plot_knn_bound(X, y, n_neighbors=1)
plot_knn_bound(X, y, n_neighbors=10)
```

```python
X_tr, Y_tr = fetch_20newsgroups(subset='train', return_X_y=True, random_state=42)
X_ts, Y_ts = fetch_20newsgroups(subset='test', return_X_y=True, random_state=42)

X_tr = np.array(X_tr, dtype=object)
X_ts = np.array(X_ts, dtype=object)
Y_tr = np.array(Y_tr)
Y_ts = np.array(Y_ts)

T_fid = TfidfVectorizer(max_df=0.8, min_df=10, max_features=1000, stop_words='english')
T_fid.fit(X_tr)
X_tr_T_fid = T_fid.transform(X_tr)
X_ts_T_fid = T_fid.transform(X_ts)

best_model = KNeighborsClassifier(n_neighbors=1, metric='cosine', weights='uniform')
best_model.fit(X_tr_T_fid, Y_tr)

y_pred = best_model.predict(X_ts_T_fid)
score = accuracy_score(Y_ts, y_pred)
print(f"accuracy    = {score}")
```

```python
def plot_distributions(train_labels, test_labels):
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.histplot(train_labels, kde=False)
    plt.title('Training sample')
    plt.xlabel("The value of the target variable", fontsize=12)
    plt.ylabel("Frequency", fontsize=12)

    plt.subplot(1, 2, 2)
    sns.histplot(test_labels, kde=False)
    plt.title('Test sample')
    plt.xlabel("The value of the target variable", fontsize=12)
    plt.ylabel("Frequency", fontsize=12)

    plt.suptitle('Distribution of training and test samples', fontsize=13)

    plt.show()

#       
plot_distributions(y_train, y_test)
```
